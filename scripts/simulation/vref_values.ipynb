```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./data.txt', sep=r'\s+', header=None)

# Flatten the data into a 1D array
flattened_data = data.values.flatten()

# Reshape the data into the desired shape (63, 2)
reshaped_data = flattened_data[:63 * 2].reshape(63, 2)  # Truncate if excess data exists

# Create a new DataFrame for the reshaped data
reshaped_df = pd.DataFrame(reshaped_data, columns=['Column1', 'Voltage References'])

real_ref = reshaped_df['Voltage References'].tolist()
```

```python
bits = 6
v_ref = 1.8

lsb = v_ref/(2**bits)

ideal_ref = []
for i in range(64):
    ideal_ref.append((i+1)*lsb)

```

```python
list1 = real_ref
list2 = ideal_ref

# Subtract list1 from list2 at each index
difference = [(b - a)*1e3 for a, b in zip(list1, list2)]
```

```python
plt.plot(difference)
```

```python
plt.plot(real_ref)
plt.plot(ideal_ref)
```
