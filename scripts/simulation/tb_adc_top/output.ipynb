```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../../../simulation/simulation/tb_adc_top.sch/ideal.txt', sep=r'\s+', header=None)

x_unit = 1e6
y_unit = 1

plt.plot(data[0]*x_unit, data[1]*y_unit)
plt.plot(data[2]*x_unit, data[3]*y_unit)

plt.xlabel("Time (us)")
plt.ylabel("Voltage (V)")
plt.title("Ideal 6-bit ADC")

plt.grid(True)
```
